## SCRAPING SHOPEE KEYWORD SEPATU VANS

```python
import requests
import csv
from google.colab import drive
import pandas as pd
import json
```

```python
drive.mount('/content/drive')
```

```python
file_simpan = 'drive/My Drive/Colab Notebooks/Homework_Web_Scraping/'
```

```python
url = 'https://shopee.co.id/api/v4/search/search_items?by=relevancy&keyword=sepatu%20vans&limit=100&newest=0&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2'

r = requests.get(url).json()
```

```python
#print(r.status_code)
```

```python
products = r['items']

header = ['itemid','nama','harga','rating']

csv_file =  open(file_simpan + 'sepatu_vans_shopee.csv', 'w')

csv_writer = csv.writer(csv_file,delimiter=",")

csv_writer.writerow(header)

for p in products:
  itemid = p['itemid']
  nama = p['item_basic']['name']
  harga = p['item_basic']['price']
  rating = p['item_basic']['item_rating']['rating_star']
  #print('item :',itemid, 'nama : ',nama, 'harga : ',harga, 'rating : ',rating)
  data = [itemid,nama,harga,rating]
  csv_writer.writerow(data)

csv_file.close()
```
